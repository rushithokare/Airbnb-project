# Data Cleaning

Load the raw Airbnb datasets and inspect their structure before cleaning.

```python
import pandas as pd

# Load data
listings = pd.read_csv("../data/raw/listings.csv")
calendar = pd.read_csv("../data/raw/calendar.csv")
reviews = pd.read_csv("../data/raw/reviews.csv")

# Check shape, head, and columns
for name, df in {
    "listings": listings,
    "calendar": calendar,
    "reviews": reviews,
}.items():
    print(f"\n{name.upper()}")
    print("Shape:", df.shape)
    display(df.head())
    print("Columns:", df.columns.tolist())
```

```python
# Clean price columns
listings['price'] = listings['price'].replace(r'[\$,]', '', regex=True).astype(float)
calendar['price'] = calendar['price'].replace(r'[\$,]', '', regex=True).astype(float)

# Convert dates
calendar['date'] = pd.to_datetime(calendar['date'])
reviews['date'] = pd.to_datetime(reviews['date'])

# Handle missing values
listings['reviews_per_month'] = listings['reviews_per_month'].fillna(0)

# Save cleaned listings
listings.to_csv("../data/processed/cleaned_listings.csv", index=False)

print('Price columns cleaned, dates converted, missing values handled, and cleaned listings saved.')
```
