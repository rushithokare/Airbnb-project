# Exploratory Data Analysis

Explore the cleaned Airbnb listings data with a few core distribution and relationship checks.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load cleaned data
listings = pd.read_csv("../data/processed/cleaned_listings.csv")

# Price distribution
sns.histplot(listings['price'], bins=50)
plt.title("Price Distribution")
plt.show()

# Room type analysis
listings.groupby('room_type')['price'].mean().plot(kind='bar')
plt.title("Price by Room Type")
plt.ylabel("Average Price")
plt.show()

# Correlation heatmap
sns.heatmap(listings.corr(numeric_only=True), cmap="coolwarm")
plt.show()
```
